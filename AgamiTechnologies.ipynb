```python
class Student:
    def __init__(self,name,roll_no,percentage):
        self.name=name
        self.roll_no=roll_no
        
        self.percentage=percentage
        

    
```

```python
def add(name,roll_no,percentage):
    obj1=Student(name,roll_no,percentage)
    print("data entered")
    return obj1
```

```python
def delete(data,name):
        for i in range(len(data)):
            if data[i].name==name:
                x=i
                break
        else:print("Student does not exist")
        del data[x]
        print("Student deleted")
```

```python
def view(data):
        print("_________________")
        if len(data)<1:
            print("No students found")
        for i in data:
            print("Student Name =",i.name,"Student Roll_No =",i.roll_no,"Student Percentage =",i.percentage,sep=" ")
        print("_________________")
```

```python
def form():
    print('Enter "Exit" to quit ')
    print('Enter "delete" to delete existing student')
    print('Enter "update" To update Existing Student')
    print('Enter "add" to enter data of student')
    print('Enter "delete+" to delete student data in bulk')
    print("Enter 'view' to show all Student List")
    print("Enter 'update+ to bulk update data in list'")
```

```python
def update(data,name):
    delete(data,name)
    name=input("Enter New Name Of Student")
    roll_no=input("Enter New Roll_NO Of Student")
    Percentage=input("Enter New Percentage Of Student")
    r=add(name,roll_no,Percentage)
    data.append(r) 
    print('Data Updated')
```

```python
data=[]
while(True):
    form()
    formresponse=input()
    #exiting
    if formresponse=="Exit":
        break
        
    #deleting in bulk
    elif formresponse=="delete+":
        
        flag=True
        while(flag):
            view(data)
            print("Enter Quit To Exit From Bulk Delete AnyTime")
            name=input("Enter Name You Want To Delete")
            if name != 'Quit':
                
                delete(data,name)
            else:flag=False 
    #delete single user
    elif formresponse=="delete":
        view(data)
        name=input("Enter Name You Want To Delete")
        delete(data,name)
        print('Data deleted')
        view(data)
    #updating single user
    elif formresponse=="update":
        view(data)
        name=input("Enter name of student to update his/her data")
        update(data,name)
    #View entire data
    elif formresponse=="view":
        view(data)
    #Adding data
    elif formresponse=="add":
        name=input("Enter Name Of Student")
        roll_no=input("Enter Roll_NO Of Student")
        Percentage=input("Enter Percentage Of Student")
        r=add(name,roll_no,Percentage)
        data.append(r)
        
    #updating bulk data
    elif formresponse=="update+":
        flag=True
        while(flag):
            view(data)
            print("Enter Quit To Exit From Bulk Update AnyTime")
            name=input("Enter name of student to update his/her data")
            if name!="Quit":
                update(data,name)
            else:flag=False
            
    
    else:print("Add correct input according to Form")
```
